### Functional Method

```python
plt.plot(x,y, "r-")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")

plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
plt.show()
```

### Object Oriented Method

```python
fig = plt.figure()


axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")

plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Larger Plot")

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

plt.subplots is just a subplot manager for plt.fig

```python
fig,axes = plt.subplots(nrows=2,ncols=3)

#axes.plot(x,y)
plt.show()
plt.tight_layout()
```

The axes object can be unpacked with tuple unpacking and when unpacked is exactly the axes1 and axes2 veriables from the example above

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

For the legend, you can specify the location in the arguments. 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label = "X Squared")
ax.plot(x,x**3, label = "X Cubed")

ax.legend() 
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = "orange", linewidth = 20, alpha = 0.5, linestyle = "--") #RGB Hex Codes are also accepted
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = "orange", linewidth = 1, alpha = 0.5, linestyle = "--", marker = "o",markersize = 10, markerfacecolor = "yellow",\
        markeredgewidth =3, markeredgecolor ="green")

ax.set_xlim(0,1)
ax.set_ylim(0,2)
plt.show()
```

### Categorical Plots

#### Dataset

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.info()
```

#### Bar Plot

The estimator argument can apply aggregate functions.
- Even your own custom functions. 

```python
sns.barplot(x = "sex", y = "total_bill" , data = tips, estimator = np.std)
plt.show()
```

#### Count Plot

```python
sns.countplot(x = "sex", data = tips)
plt.show()
```

#### Box Plot

For a box plot: X = Categorical Data & Y = Quantitative Data

```python
sns.boxplot(x = "day", y = "total_bill", data = tips, hue = "smoker")
plt.show()
```

#### Violin Plot

X = Categorical Data & Y = Quantitative Data

The "split" argument splits the violin plot into 2 distributions according to the hue. 

Box Plot vs Violin Plot

Pros: 
- The Violin plot displays more infomation.
- It also shows the distribution of the data points. 

Cons:
- A box plot is easier to read and understand the underlying data. 

```python
sns.violinplot(x = "day", y = "total_bill", data = tips, hue = "sex", split= True)
plt.show()
```

#### Strip Plot

- The Jitter argument adds noise to the plot that make it easier to see the data points.
- The dodge argument is the same as the split argument.

```python
sns.stripplot(x = "day",y = "total_bill", data = tips,jitter = True, hue = 'sex', dodge = True)
plt.show()
```

#### Swarm Plot

Combine a swarm plot with a violin plot by having the violin plot in the same cell.

- The swarm plot does not scale very well with large datasets.
- It does not show very well the number of data points.
- It also takes a lot of computational power to arrange the points.
- Moreover, not very many people are familiar with this type of plot.

```python
sns.violinplot(x = "day", y = "total_bill", data = tips)
sns.swarmplot(x = "day", y = "total_bill", data = tips, color = "black")
plt.show()
```

The catplot can recreate every other plot.

```python
sns.catplot(x = "day", y = "total_bill", data = tips, kind = "bar")
plt.show()
```

### Matrix Plots

#### Dataset

```python
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc, annot = True, cmap = "coolwarm")
plt.show()
```

```python
fp = flights.pivot_table(index = "month", columns = "year", values = "passengers")
```

```python
sns.heatmap(fp, cmap = "magma", linecolor = "black",linewidths = 1)
plt.show()
```

```python
sns.clustermap(fp,cmap = "magma", linecolor = "black", linewidths = 1, standard_scale = 1)
plt.show()
```

### Grids

Datasets

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris)
plt.show()
```

#### PairGrid

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)

plt.show()
```

#### FacetGrid

```python
f = sns.FacetGrid(data = tips,col = "time", row = "smoker" )
f.map(sns.histplot, "total_bill", kde = True)
plt.show()
```

```python
f = sns.FacetGrid(data = tips,col = "time", row = "smoker" )
f.map(plt.scatter, "total_bill", "tip")
plt.show()
```
