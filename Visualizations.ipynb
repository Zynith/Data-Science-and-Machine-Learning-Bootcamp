## <en><center>Matplotlib</center></en>

### Dataset

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

### Functional Method

```python
plt.plot(x,y, "r-")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")

plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
plt.show()
```

### Object Oriented Method

```python
fig = plt.figure()


axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")

plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Larger Plot")

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

```python
fig,axes = plt.subplots(nrows= 1,ncols=2)

#axes.plot(x,y)
```

plt.subplots is just a subplot manager for plt.fig

```python
fig,axes = plt.subplots(nrows=2,ncols=3)

#axes.plot(x,y)
plt.show()
plt.tight_layout()
```

The axes object can be unpacked with tuple unpacking and when unpacked is exactly the axes1 and axes2 veriables from the example above

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

Figure Size & DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fix,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title("Title")
ax.set_ylabel("Y")
ax.set_xlabel("X")
```

For the legend, you can specify the location in the arguments. 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label = "X Squared")
ax.plot(x,x**3, label = "X Cubed")

ax.legend() 
plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = "orange", linewidth = 20, alpha = 0.5, linestyle = "--") #RGB Hex Codes are also accepted
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = "orange", linewidth = 1, alpha = 0.5, linestyle = "--", marker = "o",markersize = 10, markerfacecolor = "yellow",\
        markeredgewidth =3, markeredgecolor ="green")

ax.set_xlim(0,1)
ax.set_ylim(0,2)
plt.show()
```

## <en><center>Seaborn Visualizations</center></en>

#### Dataset

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.info()
```

### Distribution Plots

#### Displot

```python
sns.displot(tips["total_bill"], kde = False, bins = 30)
plt.show()
```

#### Joint Plot

```python
sns.jointplot(x= 'total_bill', y= "tip", data = tips,kind = "hex",)
plt.show()
```

```python
sns.jointplot(x= 'total_bill', y= "tip", data = tips,kind = "reg",)
plt.show()
```

```python
sns.jointplot(x= 'total_bill', y= "tip", data = tips,kind = "kde",)
plt.show()
```

#### Pairplot

The Hue argument takes in a categorical column.

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = "sex", palette = "coolwarm")
plt.show()
```

#### Rugplot

```python
sns.rugplot(tips["total_bill"])
plt.show()
```

```python
sns.displot(tips["total_bill"],kde = False)
plt.show()
```

#### KDE Plot

```python
sns.kdeplot(tips["total_bill"])
plt.show()
```
